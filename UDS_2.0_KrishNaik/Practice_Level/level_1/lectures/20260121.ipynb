```python
#problem statement
#we will ask used, th eamouth they want to withdraw from the bank
# if amout is smaller than balance we will say sufficent funds
# else, infsufficent funds
balance =1000
request= int(input("Enter the amout to withdraw: "))
if request < balance:
    print("sufficient funds")
else:
    print("insufficient funds")
```

```python
#nested if else
#problem statement : based on user role, we want to to give them access to the google sheet.
#adding condition : only if user is authenticated we will check their access

#admin : delete, edit ,view
#editor: edit,view
#viewer: view

is_user_authenticated = True
print("User is authenticated")
if is_user_authenticated:
    user_role = input("Enter your role (admin/editor/viewer):")

    if user_role=="admin":
        print("full access granted: delete, edit,view")
    elif user_role=="editor":
        print("Limited access: Edit, View")
    elif user_role=="viewer":
        print("restricted access: view only")
    else:
        print("Enter the valid role")
else:
    print("Please login first.")

```
