```python
#lstrip() - to remove white spaces from the string method
name = "  Monal S  "
print(len(name))
print(name.rstrip())
print(len(name.lstrip()))
print(name.lstrip())
print(len(name.lstrip()))
print(name.strip())
print(len(name.strip()))
```

```python
#problem remove the space from the both side of the data without using the strip()
name = name.lstrip()
name= name.rstrip()
print(name)
print(len(name))
```

# if-elif-else

```python
operation_name="add"
#control flow start
if operation_name =="add":
    print("perform addition")
elif operation_name=="substract":
    print("Perform substraction")
elif operation_name=="divid":
    print("Perform division")
else:
    print("No selected operation name found")
#control flow end
```

```python
age=19
batch = 2.0
session_number = 4

#control flow start
if age > 19:
    print("19")
#control flow end

#control flow start
if batch > 1.0:
    print("give access to v1.0")
#control flow end

#control flow start
if session_number>1:
    print("inducion class ended")
#control flow end
```
